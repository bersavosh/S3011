```python
import pandas as pd
import matplotlib.pyplot as plt

DATA_DIR = 'https://raw.githubusercontent.com/bersavosh/S3011/main/data/'
```

## Our data

```python
data = pd.read_csv(DATA_DIR+'movement-speeds-hourly-san-francisco-2020-1_first4k.csv')
data
```

## Making a simple plot

Implict

```python
plt.plot(data['speed_mph_mean'])
```

Explicit

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()
ax.plot(data['speed_mph_mean'])
```

Now customizing the plot

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()
ax.plot(data['speed_mph_mean'], ':s', markersize=2, linewidth=0.5, color='green')
ax.set_xlabel('Trip Number')
ax.set_ylabel('Mean speed (mph)')
```

## Saving: raster versus vector rendering

```python
fig.savefig('plot_raster.jpg')
fig.savefig('plot_vector.pdf')
```
